```python
import pandas as pd
import matplotlib as plt
import numpy as np
from scipy.stats import norm
% matplotlib inline
```

```python
apl = pd.read_csv('../data/apple.csv', index_col=0)
apl.plot()
```

```python
apl['logreturn'] = np.log(apl['Close'].shift(-1)) - np.log(apl['Close'])
xbar = apl['logreturn'].mean()
s = apl['logreturn'].std(ddof=1)
n = apl['logreturn'].shape[0]
zhat = (xbar - 0) / (s/(n**0.5))
print(zhat)
```

```python
alpha = 0.05
z_left = norm.ppf(alpha / 2, 0, 1)
z_right = -z_left
print(z_left, z_right)
print(f"At the significance level of {alpha}")
print(f"Shall we reject? {zhat > z_right or zhat < z_left}")
```

```python
# p value for two tails test
alpha = 0.05
p = 1 - (norm.cdf(abs(zhat), 0, 1))
print(f"At the significance level of {alpha}")
print(f"Shall we reject? {p < alpha}")
```
